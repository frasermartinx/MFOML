```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
l=1
def exponentiated_quadratic(xa,xb,l,sig):
    #squared distance
    sq_norm = scipy.spatial.distance.cdist(xa,xb,"sqeuclidean")
    return sig * np.exp(-1/(2*l)*sq_norm)
```

```python
#sample from gaussian process
N = 41
nfuncs = 5

#vary
sig = 10
l = 1


X = np.expand_dims(np.linspace(-4,4,N),1) #make it a vector
cov = exponentiated_quadratic(X,X,l,sig) # kernel
#draw samples from prior as data points
#mean 0
ys = np.random.multivariate_normal(mean = np.zeros(N), cov = cov, size = 1)
```

```python
#plot
plt.figure(figsize = (6,4))
plt.plot(X,ys.T, linestyle = "-", marker = "o", markersize = 3)
plt.xlim([-4,4])
plt.show()
```
